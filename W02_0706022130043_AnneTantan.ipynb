```python
# Fill the Nan column with median
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())
df['land_size_m2'] = df['land_size_m2'].fillna(df['land_size_m2'].median())
df['building_size_m2'] = df['building_size_m2'].fillna(df['building_size_m2'].median())
df['floors'] = df['floors'].fillna(df['floors'].median())
df['building_age'] = df['building_age'].fillna(df['building_age'].median())
## Filling missing numeric values with the median is a robust approach because it is less influenced by extreme outliers compared to the mean, making it a more reliable representation of central tendency.
## This ensures the dataset remains consistent and minimizes bias, especially in real estate data where values like land size or price can vary widely.
```

```python
# Fill the Nan values in certificate and building_orientation with "Unknown", because because they are categorical features that cannot be filled using numerical methods.
# This approach preserves all rows, avoids data loss, and clearly indicates missing or unprovided information.
df['certificate'] = df['certificate'].fillna("Unknown")
df['building_orientation'] = df['building_orientation'].fillna("Unknown")
```

```python
#Binning building_age into categorical ranges
bins = [-1, 5, 10, 15, 20, 25, float('inf')]
labels = ['0-5 years', '6-10 years', '11-15 years', '16-20 years', '21-25 years', '> 25 years']

df['building_age'] = pd.cut(
    df['building_age'],
    bins=bins,
    labels=labels
)
```

```python
import numpy as np

# Step 1: If 'furnishing' has the value 'baru', replace it with 'Unknown'
# Reason: 'baru' is inconsistent with furnishing categories,
# so we standardize it as 'Unknown' to avoid confusion.
df['furnishing'] = df['furnishing'].replace('baru', 'Unknown')

# Step 2: Fill missing values in 'furnishing' using rules from 'property_condition'
# Reason: If property_condition is 'semi furnished' or 'unfurnished',
# we can assume furnishing follows the same. Otherwise, mark as 'Unknown'.
df.loc[df['furnishing'].isna() & df['property_condition'].isin(['semi furnished', 'unfurnished']), 'furnishing'] = df['property_condition']
df.loc[df['furnishing'].isna() & ~df['property_condition'].isin(['semi furnished', 'unfurnished']), 'furnishing'] = 'Unknown'

# Step 3: Fill missing values in 'property_condition' using rules from 'furnishing'
# Reason: If furnishing is 'baru', copy that value into property_condition.
# Otherwise, mark it as 'Unknown' since no reliable mapping is available.
df.loc[df['property_condition'].isna() & (df['furnishing'] == 'baru'), 'property_condition'] = df['furnishing']
df.loc[df['property_condition'].isna() & (df['furnishing'] != 'baru'), 'property_condition'] = 'Unknown'

# Step 4: Standardize 'property_condition' by replacing 'semi furnished' and 'unfurnished' with 'Unknown'
# Reason: To simplify categories and reduce sparsity by grouping minor/uncertain categories as 'Unknown'.
df['property_condition'] = df['property_condition'].replace(['semi furnished', 'unfurnished'], 'Unknown')
```

```python
# Mapping scores
furnishing_scores = {
    'Unknown': 0,
    'unfurnished': 0,
    'semi furnished': 1,
    'furnished': 2
}

property_condition_scores = {
    'Unknown': 0,
    'needs renovation': 0,
    'good': 1,
    'very good': 2,
    'renovated': 2,
    'new': 3
}

# Create score columns
df['furnishing_score'] = df['furnishing'].map(furnishing_scores)
df['property_condition_score'] = df['property_condition'].map(property_condition_scores)

# Create new feature (Feature Engineering)
# Total score = sum of both
df['house_condition'] = df['furnishing_score'] + df['property_condition_score']
# Convert categorical values into numerical scores so they can be compared, analyzed, or used in models.
# Summing the furnishing score and property condition score creates a single house_condition metric, making it easier to evaluate overall property quality.
```
